## **ðŸ”¸ Logistic Regression (sklearn)**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier

from sklearn.datasets import load_breast_cancer, make_classification, make_blobs, make_circles
```

```python
X, y = load_breast_cancer(return_X_y=True)
X.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

### `LogisticRegression()`

```python
LogisticRegression()
```

```python
# model = LogisticRegression()
model = LogisticRegression(solver='sag', max_iter=200, random_state=14)
model.fit(x_train, y_train)
model.predict(x_test), y_test
```

```python
model.predict_proba(x_test)
# model.predict_proba(x_test).sum(1)
```

```python
model.predict_log_proba(x_test)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

### My test: Using class_weight

```python
model = LogisticRegression(solver='sag', max_iter=200,class_weight="balanced", random_state=14)
model.fit(x_train, y_train)
model.predict(x_test)
model.score(x_train, y_train)
model.score(x_test, y_test)
```

```python
model = LogisticRegression(solver='sag', max_iter=200,class_weight={0:0.1,1:0.2,3:1}, random_state=14)
model.fit(x_train, y_train)
model.predict(x_test)
model.score(x_train, y_train)
model.score(x_test, y_test)
```

```python
import numpy as np

print(x_test.shape[1])
feature_weights = {i: np.random.uniform(0.0, 1.0) for i in range(x_test.shape[1])}
feature_weights.update({0: 0.1, 1: 0.2, 3: 1.0})
# feature_weights[0] = 0.2
# feature_weights[1] = .1
# # feature_weights[3] = 1
print(feature_weights)
model = LogisticRegression(
    solver="sag", max_iter=200, class_weight=feature_weights, random_state=14
)
model.fit(x_train, y_train)
model.predict(x_test)
model.score(x_train, y_train)
model.score(x_test, y_test)
```

### `SGDClassifier()`

```python
model = SGDClassifier(loss='log_loss', random_state=27)
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model = SGDClassifier(loss='hinge', random_state=27)
model.fit(x_train, y_train)
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```
