```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

import os

notebook_path = os.getcwd()
```

```python
# Datasets path
# auto
csv_train_path = f"{notebook_path}/data/auto-train-multi.csv"
csv_test_path = f"{notebook_path}/data/auto-test-multi.csv"
```

```python
df=pd.read_csv(csv_train_path)
train_set=np.array(df)
# df

x_train=train_set[:,:-1]
y_train=train_set[:,-1:]
x_train.shape,y_train.shape
```

```python
df=pd.read_csv(csv_test_path)
test_set=np.array(df)
# df

x_test=test_set[:,:-1]
y_test=test_set[:,-1:]
x_test.shape,y_test.shape
```

```python
# Visualize
i=0
plt.scatter(x_train[:,i],y_train)
plt.scatter(x_test[:,i],y_test)
```

```python
d=2
```

```python
poly= PolynomialFeatures(degree=d)
phi_x_train=poly.fit_transform(x_train)
```

```python
wc = np.linalg.inv(phi_x_train.T @ phi_x_train).T @ phi_x_train.T @ y_train
```

```python
y_hat_train=  phi_x_train @ wc
r2s_train=r2_score(y_train, y_hat_train)
```

```python
phi_x_test=poly.transform(x_test)
y_hat_test=  phi_x_test @ wc
r2s_test=r2_score(y_test, y_hat_test)
```
