```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin_x_plus_y(x, y):#sin(x+y)
    return np.sin(x + y)


def grad_sin_x_plus_y(x, y):
    return np.cos(x) , np.cos(y)

def sin_x_mul_y(x, y): #sin(x*y)
    return np.sin(x * y)


def grad_sin_x_mul_y(x, y):
    return y*np.cos(x*y) , x*np.cos(x*y)



def f3d(x, y):
    return x**2 + y**2


def grad_f3d(x, y):
    return 2 * x , 2 * y


def gradient_dissent(grad_func, xi, yi, eta, epochs):
    x_min=xi
    y_min=yi
    for i in range(epochs):
        grad=grad_func(x_min,y_min)
        x_min=x_min-eta*grad[0]
        y_min=y_min-eta*grad[1]

    return x_min,y_min
        
```

```python
x, y = np.meshgrid(np.linspace(0, 3), np.linspace(0, 3))

# func_ = f3d
# grad_func_ = grad_f3d

# func_=sin_x_plus_y
# grad_func_=grad_sin_x_plus_y

func_=sin_x_mul_y
grad_func_=grad_sin_x_mul_y

xi = 1.5
yi = 1.5
eta = 0.1
epochs = 20
x_min,y_min= gradient_dissent(grad_func_, xi, yi, eta, epochs)


z = func_(x, y)
ax = plt.axes(projection="3d")
ax.plot_surface(x, y, z, cmap="viridis", alpha=0.6)

ax.scatter(xi,yi,func_(xi,yi),color='r', s=50, zorder=2, label='Start Point')
ax.scatter(x_min,y_min,func_(x_min,y_min),color='k', s=50, zorder=2, label='End Point')
plt.legend();
# ax.plot(gdl[0],gdl[1],,func_(x0,y0))

# fig = plt.figure(figsize=(10, 6))
# ax = fig.add_subplot(111, projection='3d')     # for 3d plot
# ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6);
# ax.scatter(x1,y1,func_(x1,y1), color='k', s=50, zorder=2, label='End Point')
# ax.scatter(x0,y0,func_(x0,y0), color='r', s=50, zorder=2, label='Start Point')
# plt.legend();
```
