```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn_f(x):
    return np.sin(3*x)
def fn_fp(x):
    return 3*np.cos(3*x)
```

```python
# x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
x_main = np.linspace( -np.pi/2, np.pi/2, 1000)

y_main = fn_f(x_main)
yp_main = fn_fp(x_main)
```

```python
lr=0.1
x0=-1.5
epochs=20

x_grad=x0
for epoch in range(epochs):
    grad= fn_fp(x_grad)
    x_grad=x_grad-lr*grad
    y_grad=fn_f(x_grad)
print(x_grad,y_grad)
```

```python

plt.plot(x_main, y_main, color="red")
plt.plot(x_main, yp_main, color="blue")
# plt.plot(x_grad, y_grad, color="green")
plt.scatter(x_grad, y_grad, color="green")

m=fn_fp(x_grad)
yyyy= m*(x_main-x_grad)+y_grad
plt.plot(x_main, yyyy, color="green")

plt.show()
```
