```python
# Linear regression function
def fn_linear_regression(x, w0, w1):
    return w1 * x + w0
```

```python
# MAE
def fn_mae(y, yhat):
    return np.mean(np.abs(y - yhat))
```

```python
# MSE
def fn_mse(y, yhat):
    return np.mean((y - yhat)**2)
```

```python
x = np.linspace(-5, 5, 100)
y = 2 * x + 1

w0 = 0
wl1 = np.linspace(-100, 100, 1000)

el_mse = []
el_mae = []

for w1 in wl1:
    yhat = fn_linear_regression(x, w0, w1)
    e = fn_mae(y, yhat)
    el_mae.append(e)
    e = fn_mse(y, yhat)
    el_mse.append(e)
el_mse = np.array(el_mse)
el_mae = np.array(el_mae)

# plt.scatter(wl1, el_mae)
plt.plot(wl1, el_mae,"--",color="blue")
plt.show()
plt.plot(wl1, el_mse,"--",color="red")
plt.show()
```
